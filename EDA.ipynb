# Preprocessing Data

Menggabungkan data tahunan dari 2014 sampai 2023, lalu mengambil hanya data yang memiliki kode SITC 321.

```python
dataframes = [
    (df_2014, 2014),
    (df_2015, 2015),
    (df_2016, 2016),
    (df_2017, 2017),
    (df_2018, 2018),
    (df_2019, 2019),
    (df_2020, 2020),
    (df_2021, 2021),
    (df_2022, 2022),
    (df_2023, 2023)
]

data_batubara = []

for df, tahun in dataframes:
    df_321 = df[df.iloc[:, 0].str.startswith("321", na=False)].copy()
    df_321["Tahun"] = tahun
    data_batubara.append(df_321)

data_321 = pd.concat(data_batubara, ignore_index=True)

data_321.rename(columns={
    'Unnamed: 1': 'Januari',
    'Unnamed: 2': 'Februari',
    'Unnamed: 3': 'Maret',
    'Unnamed: 4': 'April',
    'Unnamed: 5': 'Mei',
    'Unnamed: 6': 'Juni',
    'Unnamed: 7': 'Juli',
    'Unnamed: 8': 'Agustus',
    'Unnamed: 9': 'September',
    'Unnamed: 10': 'Oktober',
    'Unnamed: 11': 'November',
    'Unnamed: 12': 'Desember',
    'Unnamed: 13': 'Rata-Rata Tahunan'
}, inplace=True)

data_321
```

Menghapus kolom yang tidak diperlukan dalam analisis.

```python
data_321.drop(columns=['Kode SITC 3 digit (2018=100)', 'Rata-Rata Tahunan'], inplace=True)
```

# EDA

## Data Preparation

 Melakukan transformasi bentuk data dari format lebar (wide) ke format panjang (long), agar lebih mudah dianalisis per bulan.

```python
data = pd.melt(data_321, id_vars='Tahun', var_name='Bulan', value_name='Indeks')
data
```

 Menyusun ulang nama bulan menjadi format angka (mis. "Januari"  "01") agar dapat dibuat kolom tanggal. Kemudian menambahkan kuartar untuk mengetahui pola musiman.

```python
bulan_mapping = {
    'Januari': '01', 'Februari': '02', 'Maret': '03', 'April': '04',
    'Mei': '05', 'Juni': '06', 'Juli': '07', 'Agustus': '08',
    'September': '09', 'Oktober': '10', 'November': '11', 'Desember': '12'
}
data['Tanggal'] = pd.to_datetime(data['Tahun'].astype(str) + '-' + data['Bulan'].map(bulan_mapping) + '-01')
data['Kuartal'] = 'Q' + data['Tanggal'].dt.quarter.astype(str)


data
```

Menyimpan hasil data yang telah dipersiapkan ke dalam file CSV untuk lanjut ke tahap modeling.

```python
data.to_csv('data.csv', index=False)
```

Mengatur tipe data untuk kolom Tahun, Bulan, dan Indeks agar sesuai untuk analisis dan manipulasi lebih lanjut.

```python
data['Tahun'] = data['Tahun'].astype(str)
data['Bulan'] = data['Bulan'].astype(str)
data['Kuartal'] = data['Kuartal'].astype(str)
data['Indeks'] = data['Indeks'].astype(float)
```

## Statistical Descriptive

Menampilkan ringkasan statistik.

```python
data.describe(include="all")
```

Menampilkan ringkasan statistik seperti mean, standar deviasi, nilai minimum, maksimum, dan kuartil.

```python
data['Indeks'] = pd.to_numeric(data['Indeks'], errors='coerce')

month_order = ['Januari', 'Februari', 'Maret', 'April', 'Mei', 'Juni',
               'Juli', 'Agustus', 'September', 'Oktober', 'November', 'Desember']

yearly_stats = data.groupby('Tahun')['Indeks'].agg(['min', 'max', 'mean', 'median', 'std'])
yearly_stats = yearly_stats.rename(columns={
    'min': 'Minimum',
    'max': 'Maximum',
    'mean': 'Mean',
    'median': 'Median',
    'std': 'Standard Deviation'
})
yearly_stats['Percentage Change'] = yearly_stats['Mean'].pct_change() * 100
yearly_stats['Coefficient of Variation'] = yearly_stats['Standard Deviation'] / yearly_stats['Mean']

monthly_stats = data.groupby('Bulan')['Indeks'].agg(['min', 'max', 'mean', 'median', 'std'])
monthly_stats = monthly_stats.rename(columns={
    'min': 'Minimum',
    'max': 'Maximum',
    'mean': 'Mean',
    'median': 'Median',
    'std': 'Standard Deviation'
})
monthly_stats['Coefficient of Variation'] = monthly_stats['Standard Deviation'] / monthly_stats['Mean']
monthly_stats['Percentage Change'] = monthly_stats['Mean'].pct_change() * 100

monthly_stats.index = pd.CategoricalIndex(monthly_stats.index, categories=month_order, ordered=True)
monthly_stats = monthly_stats.sort_index()

print("Yearly Statistics:\n", yearly_stats)
print("\nMonthly Statistics:\n", monthly_stats)
```

# Kesimpulan dan Saran

## Kesimpulan

Pasar batubara Indonesia selama dekade terakhir ditandai oleh **fase penurunan, pemulihan, shock pandemi, lonjakan krisis energi, dan koreksi pasar**. Meskipun ada fluktuasi ekstrem, **pola musiman tetap konsisten** dan dapat dimanfaatkan dalam perencanaan strategi bisnis.
**Volatilitas pasar** tetap menjadi tantangan utama yang memerlukan perencanaan jangka panjang dan strategi manajemen harga yang adaptif.

**Tren Jangka Panjang (Tahunan)**

- 2014-2016: Periode penurunan harga signifikan hingga level 55-60 akibat kelebihan pasokan global dan perlambatan ekonomi, terutama karena ketergantungan Indonesia terhadap pasar Cina.
- 2017-2019: Pemulihan stabil dengan harga kembali ke level 85-100 seiring membaiknya ekonomi global.
- 2020-2021: Pandemi COVID-19 menyebabkan penurunan tajam permintaan dan harga hingga menyebabkan indeks turun ke 60-65.
- 2021-2022: Lonjakan harga ekstrem dipicu oleh krisis energi global (konflik Rusia-Ukraina) dan gangguan rantai pasok; 2022 mencatat puncak harga tertinggi (hingga sekitar 255) sepanjang dekade.
- 2023: Terjadi koreksi harga, meskipun tetap di level yang jauh lebih tinggi dibandingkan era pra-pandemi yakni seiktar 120-125.

**Volatilitas dan Siklus Pasar**

- Volatilitas tertinggi secara proporsional terjadi pada 2021 (CV  0.34), bahkan lebih tinggi dibandingkan 2022.
- Secara umum, pasar batubara sangat siklikal, dipengaruhi oleh faktor ekonomi global, geopolitik, dan gangguan pasokan.

**Pola Musiman**

- Ada siklus musiman yang stabil dan dapat diprediksi:
 - Q4 (Oktober-Desember) mencatat harga rata-rata tertinggi (115) karena meningkatnya permintaan energi menjelang musim dingin.
 - Q3 (Juli-September) menunjukkan harga rata-rata terendah (100) karena penurunan aktivitas industri dan pergeseran ke energi terbarukan.
 - Q1 & Q2 relatif stabil di kisaran 105-108.

- November-Desember bulan dengan harga rata-rata bulanan tertinggi.
- Pola musiman tetap berlaku meskipun terjadi volatilitas ekstrem di level tahunan.

**Distribusi dan Outliers**

- Boxplot menunjukkan peningkatan variabilitas harga dari awal hingga akhir tahun, dengan lebih banyak outliers ekstrem (>200) pada paruh kedua tahun.

## **Saran Kebijakan dan Strategi Pelaku Usaha Batubara**

### **Saran untuk Pemerintah**

**Perhatikan Siklus Musiman Harga**
- Data menunjukkan bahwa harga cenderung **tinggi di Q4** (Oktober-Desember) dan **rendah di Q3** (Juli-September).
